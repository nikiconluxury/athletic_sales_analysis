```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

data_2020 = pd.read_csv("./Resources/athletic_sales_2020.csv")
data_2021 = pd.read_csv("./Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_data = pd.concat([data_2020, data_2021], ignore_index=True)
```

```python
combined_data.to_csv("mydata.csv")
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_data['invoice_date'] = pd.to_datetime(combined_data['invoice_date'])
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
wf =wf.groupby(['retailer','region', 'state', 'city'])['units_sold'].sum().reset_index()
# Show the top 5 results.

wf.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
wf.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

wf.pivot_table(values='units_sold', index=['retailer','region', 'state', 'city'], aggfunc='sum')
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Show the top 5 results.

wf.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
wf.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pp = wf.pivot_table(values='total_sales', index='invoice_date').reset_index()
# Optional: Rename the "total_sales" column to "Total Salesset_index('invoice_date')
pd.to_datetime(pp['invoice_date'])
# Show the table.
pp
```

```python
pp['invoice_date'] = pd.to_datetime(pp['invoice_date'])
```

```python
pp.set_index('invoice_date', inplace=True)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales = pp.resample('D').sum()
daily_sales_sorted = daily_sales.sort_values(by='total_sales', ascending=True)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pp.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly = weekly_sales.sort_values(by='total_sales', ascending=True)
```
